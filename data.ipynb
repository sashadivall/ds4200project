```python
import pandas as pd
```

```python
df = pd.read_csv("ai_vs_human_code_dataset.csv")
```

```python
df.describe()
```

```python
import altair as alt
```

```python
alt.Chart(df).mark_boxplot().encode(
    alt.Y("complexity_score:Q", title="Complexity Score", scale=alt.Scale(padding=10)),
    alt.X("ai_generated:N", title="AI Generated?"),
    color=alt.Color("ai_generated:N", title="AI Generated")
).properties(
    title="Complexity Score for AI and Human Generated Code"
)
```

```python
hist1 = alt.Chart(df).mark_bar().encode(
    alt.X("execution_time:Q", title="Execution Time", bin=alt.Bin(extent=[0, 12], step=1)),
    alt.Y("count()", title="Count"),
    alt.ColorValue("blue")
).transform_filter(
    alt.datum.ai_generated == 0
).properties(
    title="Human"
)

hist2 = alt.Chart(df).mark_bar().encode(
    alt.X("execution_time:Q", title="Execution Time", bin=alt.Bin(extent=[0, 12], step=1)),
    alt.Y("count()", title="Count"),
    alt.ColorValue("orange")
).transform_filter(
    alt.datum.ai_generated == 1
).properties(
    title="AI"
)

(hist1 | hist2).resolve_scale(y='shared')
```
